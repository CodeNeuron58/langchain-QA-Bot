```python
# %pip install -q langchain pypdf

from langchain.document_loaders import PyPDFLoader

loader = PyPDFLoader("Rollo_Tomassi.pdf")
docs = loader.load()
```

```python
# text_splitters

from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    # Set a really small chunk size, just to show.
    chunk_size = 1000,
    chunk_overlap  = 200,
    length_function = len,
)

chucked_texts = text_splitter.split_documents(docs)

print(len(chucked_texts))
```

```python
print(chucked_texts[0].page_content)
```

```python
import os
from dotenv import load_dotenv

# Load environment variables from a local .env file (if present)
load_dotenv()

# Retrieve and validate Cohere API key from environment variables
cohere_api_key = os.getenv("COHERE_API_KEY")
if not cohere_api_key:
    raise ValueError("COHERE_API_KEY not found in .env file or environment")
```

```python
# define llm

from langchain_cohere import ChatCohere

llm = ChatCohere(
    cohere_api_key=cohere_api_key,
    model="command-a-03-2025",  # model selection can be changed
)
```

```python
# test llm

print(llm.invoke("What is ai , in one line?"))
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embedding = HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
text = "LangChain makes working with LLMs easier."
embedding_vector = embedding.embed_query(text)
print(embedding_vector)
```

```python
# vector store
from langchain_chroma import Chroma
vector_db = Chroma.from_documents(chucked_texts, embedding)
```

```python
# retriever

retriever = vector_db.as_retriever()
```

```python
from langchain.chains import RetrievalQA

qa = RetrievalQA.from_chain_type(llm, retriever=retriever)
response = qa.invoke("Whats the Title of this book? and who is the writer?")
print(response)
```
